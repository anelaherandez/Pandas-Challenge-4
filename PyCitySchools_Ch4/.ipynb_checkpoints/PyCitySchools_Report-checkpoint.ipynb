# Analysis 
   In an effort to better understand the outcome of standardized testing throughout this area, data scientists conducted analysis on the math and reading scores for each student from 15 schools (7 district schools and 8 charter schools). A variety of computations were performed to gain insight into what factors impact the overall passing of the standardized tests to improve passing rates in the future. This analysis highlighted the success seen in both math and reading testing at charter schools, as well as showed the areas of concern, particularly with regards to district schools and the overall passing rate for schools with a student population larger than 2000. 
The top five performing schools are all charter schools with student populations less than 2500 students and on average spent 620 dollars per student on test preparation. Additionally, these schools had an average 90% overall passing rate. Conversely, the bottom five schools were all district schools with student populations greater than 2500 individuals. Although these bottom schools showed a relatively solid reading passing percentage, all five schools had a less than 70% pass rate for the math test and overall failed the complete standardized test with an average passing rate of 53%. This analysis brought into question whether spending per student could potentially impact these passing outcomes, however the data does not strongly support one way or another the success of a school. The spending analysis shows that the largest category spent per student ($645-$680) had the lowest overall passing percentage (53.52% passing). The data suggests that school size is a significant factor in the success of students on passing the standardized tests. The largest category of school type (population of 2000-5000 students) showed that students only 69% of students passed the math portion of the tests and overall, only 58% passed the entire test. The schools with less than 2000 students have a much higher passing percentage in math and reading as well as overall. 
The conclusions drawn from this analysis point to a need to focus more attention to student preparation on the math portion of the test in larger district schools. For all 15 schools the reading portion of the test saw positive passing percentages (only one school had lower than 80% passing in reading), while the math scores and passing rates were quite a bit lower. The charter schools do not suffer from the same low math scores, which correlate to school size, however this attribute was not analyzed further. To better understand how these district schools can improve their math scores further analysis on classroom size, student attributes, and a more intensive look at the difference between charter and district school frameworks needs to be conducted.

```python
import pandas as pd
from pathlib import Path
```

```python
# File load
find_school_data=Path("Resources/schools_complete.csv")
find_student_data=Path("Resources/students_complete.csv")

# School and Student Data into dataframes
school_data=pd.read_csv(find_school_data)
student_data=pd.read_csv(find_student_data)

# Combine data into one dataset
Complete_School_Data=pd.merge(school_data, student_data, how="left", on=["school_name", "school_name"])
Complete_School_Data
```

# District Summary

```python
# calculate unique schools
school_count=Complete_School_Data["school_name"].unique()
School_Total=len(school_count)
School_Total
```

```python
# Total number of students
student_count=len(Complete_School_Data)
student_count
```

```python
# Total Budget by school
total_budget=school_data['budget'].sum()
total_budget
```

```python
# Calculate average math score
average_math=Complete_School_Data['math_score'].mean()
average_math
```

```python
# Calculate average reading score
average_reading=Complete_School_Data['reading_score'].mean()
average_reading
```

```python
# Calculate Percentage of students who passed math
math_pass_count=Complete_School_Data[(Complete_School_Data['math_score']>=70)].count()["student_name"]
math_pass_percentage=math_pass_count/float(student_count)*100
math_pass_percentage
```

```python
# Calculate Percentage of students who passed reading
read_pass_count=Complete_School_Data[(Complete_School_Data['reading_score']>=70)].count()['student_name']
read_pass_percentage=read_pass_count/float(student_count)*100
read_pass_percentage
```

```python
# Calculate Percentage of students that passed both subjects
pass_both_count=Complete_School_Data[(Complete_School_Data['math_score']>=70) & (Complete_School_Data['reading_score']>=70)
                                     ].count()["student_name"]

overall_pass_rate=pass_both_count/float(student_count)*100
overall_pass_rate
```

```python
# Create a high-level snapshot of the district's key metrics in a dataframe
district_summary=pd.DataFrame({"Total Schools":[School_Total], "Total Students":[student_count],"Total Budget":[total_budget],
                              "Average Math Score":[average_math], "Average Reading Score":[average_reading], 
                              "Percentage of Passing Math Scores":[math_pass_percentage], "Percentage of Passing Reading Scores":[read_pass_percentage],
                              "Overall Passing Rate":[overall_pass_rate]})
district_summary.head()
```

```python
# Formatting
district_summary["Total Students"]=district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"]=district_summary["Total Budget"].map("${:,.2f}".format)
district_summary["Average Math Score"]=district_summary["Average Math Score"].map("{:.2f}".format)
district_summary["Average Reading Score"]=district_summary["Average Reading Score"].map("{:.2f}".format)
district_summary["Percentage of Passing Math Scores"]=district_summary["Percentage of Passing Math Scores"].map("{:.2f}%".format)
district_summary["Percentage of Passing Reading Scores"]=district_summary["Percentage of Passing Reading Scores"].map("{:.2f}%".format)
district_summary["Overall Passing Rate"]=district_summary["Overall Passing Rate"].map("{:.2f}%".format)

# Display Dataframe
district_summary
```

```python
# Calculate the number of students per school with math scores of 70 or higher
students_pass_math = Complete_School_Data[Complete_School_Data["math_score"] >= 70].groupby("school_name").size()
students_pass_math
```

```python
# Calculate the number of students per school with reading scores of 70 or higher
pass_read_students=Complete_School_Data[Complete_School_Data['reading_score']>=70].groupby("school_name").size()
pass_read_students
```

```python
# Use the provided code to calculate the number of students per school that passed both math and reading with scores of 70 or higher
students_passing_math_and_reading = Complete_School_Data[
    (Complete_School_Data["reading_score"] >= 70) & (Complete_School_Data["math_score"] >= 70)
]
school_students_passing_math_and_reading=students_passing_math_and_reading.groupby(["school_name"]).size()
```

```python
# Use the provided code to calculate the passing rates
Math_success_school=students_pass_math/Student_per_School_count*100
Read_success_school=pass_read_students/Student_per_School_count*100
overall_passing_rate=school_students_passing_math_and_reading/Student_per_School_count*100
overall_passing_rate
```

```python
# School Types
Type_of_School=school_data.set_index(["school_name"])['type']
Type_of_School                                                    
```

```python
# Create a Data Frame of a summary of all Per School Calculations
school_summary=pd.DataFrame({"School Type":Type_of_School, "Total Students": Student_per_School_count,
                            "Total School Budget": school_budget, "Per Student Budget": School_per_capita,
                            "Average Math Score":average_math_school, "Average Reading Score": average_read_school,
                            "% Passing Math": Math_success_school, "% Passing Reading": Read_success_school,
                            "% Overall Passing": overall_passing_rate})
school_summary
```

## Scores by School Spending

```python
# Establish the bins
spend_bin=[0,585,630,645,680]
labels=["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Copy of School Summary Data Frame
school_spend_df=school_summary.copy()
```

```python
# use pd.cut to categorize spending based on bins
school_spend_df["Spending Ranges (Per Student)"]=pd.cut(school_spend_df["Per Student Budget"], spend_bin, labels=labels,
                                                       include_lowest=True)
school_spend_df
```

```python
#  Calculate averages for the desired columns. 
spending_math_scores = school_spend_df.groupby(["Spending Ranges (Per Student)"])["Average Math Score"].mean()
spending_reading_scores = school_spend_df.groupby(["Spending Ranges (Per Student)"])["Average Reading Score"].mean()
spending_passing_math = school_spend_df.groupby(["Spending Ranges (Per Student)"])["% Passing Math"].mean()
spending_passing_reading = school_spend_df.groupby(["Spending Ranges (Per Student)"])["% Passing Reading"].mean()
overall_passing_spending = school_spend_df.groupby(["Spending Ranges (Per Student)"])["% Overall Passing"].mean()
spending_math_scores
```

```python
# Create spending data frame
Spend_summary=pd.DataFrame({"Average Math Score": spending_math_scores, "Average Reading Score": spending_reading_scores,
                          "% Passing in Math": spending_passing_math, "% Passing in Reading": spending_passing_reading,
                            "% Overall Passing": overall_passing_spending})
Spend_summary
```

## Scores by School Size

```python
# establish bin and label
size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
# Use `pd.cut` on the "Total Students" column of the `per_school_summary` DataFrame.
school_summary["School Size"]=pd.cut(school_summary["Total Students"], size_bins, labels=labels, include_lowest=True)
school_summary
```

```python
# Calculate averages for the desired columns. 
size_math_scores = school_summary.groupby(["School Size"])["Average Math Score"].mean()
size_reading_scores = school_summary.groupby(["School Size"])["Average Reading Score"].mean()
size_passing_math = school_summary.groupby(["School Size"])["% Passing Math"].mean()
size_passing_reading = school_summary.groupby(["School Size"])["% Passing Reading"].mean()
size_overall_passing = school_summary.groupby(["School Size"])["% Overall Passing"].mean()
```

```python
# Create a DataFrame called `size_summary` that breaks down school performance based on school size (small, medium, or large).
size_summary=pd.DataFrame({"Average Math Score": size_math_scores, "Average Reading Score": size_reading_scores,
                          "% Passing in Math": size_passing_math, "% Passing in Reading": size_passing_reading,
                          "% Overall Passing": size_overall_passing})
size_summary
```

```python
# Group the per_school_summary DataFrame by "School Type" and average the results.
average_math_score_by_type = school_summary.groupby(["School Type"])["Average Math Score"].mean()
average_reading_score_by_type = school_summary.groupby(["School Type"])["Average Reading Score"].mean()
average_percent_passing_math_by_type = school_summary.groupby(["School Type"])["% Passing Math"].mean()
average_percent_passing_reading_by_type = school_summary.groupby(["School Type"])["% Passing Reading"].mean()
average_percent_overall_passing_by_type = school_summary.groupby(["School Type"])["% Overall Passing"].mean()
```

```python
# Data Frame by School type
type_summary=pd.DataFrame({"Average Math Score": average_math_score_by_type, "Average Reading Score": average_reading_score_by_type,
                          "% Passing in Math": average_percent_passing_math_by_type, "% Passing in Reading": average_percent_passing_reading_by_type,
                          "% Overall Passing": average_percent_overall_passing_by_type})
type_summary
```

```python
# Formatting Clean up of School Type Summary
type_summary["Average Math Score"]=type_summary["Average Math Score"].map("{:.2f}".format)
type_summary["Average Reading Score"]=type_summary["Average Reading Score"].map("{:.2f}".format)
type_summary["% Passing in Math"]=type_summary["% Passing in Math"].map("{:.2f}%".format)
type_summary["% Passing in Reading"]=type_summary["% Passing in Reading"].map("{:.2f}%".format)
type_summary["% Overall Passing"]=type_summary["% Overall Passing"].map("{:.2f}%".format)
type_summary
```
